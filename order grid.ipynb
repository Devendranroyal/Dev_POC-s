```python
Word Cloud is a data visualization technique used for representing text data in 
    which the size of each word indicates its frequency or importance.
This format is useful for quickly perceiving the most prominent terms.
It displays a list of words, the importance of each beeing shown with font size or color.
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
text = df1['Failed Workflow Name'].values
words = '' 
for val in df1['Failed Workflow Name']:   
     val = str(val) 
     tokens = val.split('_')  
     words += " ".join(tokens)+" "
wordcloud = WordCloud(
    width = 300,
    height = 300,
    background_color = 'black',
    min_font_size=4,
    stopwords = STOPWORDS).generate(str(words))
fig = plt.figure(
    figsize = (8, 8),
    facecolor=None)
plt.imshow(wordcloud, interpolation = 'spline36')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

## 9.Calculate sum of value of cbu(pie_data)



```python
df7=pd.read_excel('pie_data.xlsx')
df7
```

```python
div=len(df7.index)
data=pd.Series(df7.Value.values,index=df7.CBU).to_dict()
data
```

```python
output_notebook()

data1=pd.Series(data).reset_index(name='Value').rename(columns={'index':'CBU'})
#data = pd.Series(x).reset_index(name='value').rename(columns={'index':'fruit'})
data1['angle'] = data1['Value']/data1['Value'].sum() * 2*pi
data1['color'] = Set3[div]
#data1['cumulative_angle']:[(sum(value[0:i+1])- (item/2))/sum(value)*2*pi for i,item in enumerate(value)]
cumulative = data1['Value'].cumsum()/data1['Value'].sum()
cumulative_s=cumulative.copy()
cumulative_s[0] = 0
cumulative_s[1:] = cumulative[:-1]
s=cumulative_s*2*pi + 0.5*pi
e=cumulative*2*pi +0.5*pi
value=data1['Value'].tolist()
cumulative_angle=[(sum(value[0:i+1])- (item/2))/sum(value)*2*pi for i,item in enumerate(value)]

data1['cumulative_s']=s
data1['cumulative']=e
data1['cos'] = np.cos(cumulative_angle)*0.4

data1['sin'] = np.sin(cumulative_angle)*0.4
sep = []
for i in range(len(data1.index)):
    sep.append(': ')
data1['legend'] = data1['CBU'] + sep + data1['Value'].astype(str)
data1['label'] = ["{:.0f}".format(l) for l in data1['Value']]
data1['label'] = data1['label'].astype(str)
plot4 = figure(title="Value of CBU", toolbar_location='right',
          tooltips="@CBU: @Value",x_range=(-0.5,1.5))
source = ColumnDataSource(data1)
plot4.annular_wedge(x=0, y=0,inner_radius=0.1,outer_radius=0.5,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='legend', source=data1)
labels = LabelSet(x='cos',y='sin',text='label', source=source,angle=0,
                  text_font_size="10pt", text_color="black",text_align='center')
plot4.add_layout(labels,'below')
#p.add_layout(legend,'right')
plot4.axis.axis_label=None
plot4.axis.visible=False
plot4.xgrid.visible=False
plot4.ygrid.visible=False
#show(plot4)
#print(cumulative,cumulative_s,s,e)
```
