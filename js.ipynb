```python
from bokeh.io import show
from bokeh.models import ColumnDataSource,LabelSet
from bokeh.palettes import Category10
from bokeh.plotting import figure,output_notebook
from bokeh.models.tools import HoverTool



nn=df_val_counts['type'].tolist()
mm=df_val_counts['count'].tolist()

source = ColumnDataSource(dict(
    x=[2, 4, 2.7, 3.55, 3.2],
    y=[2, 1.3, 0.7, 2.5, 2],
    color=Category10[5],
    label=nn,
    radius= [0.9,0.8,0.6,0.4,0.2],
    v=nn,
    c=mm,
    names=nn
))

p = figure(x_range=(0, 7), y_range=(0, 3), plot_height=300, tools="zoom_in,zoom_out,save,reset")


# legend field matches the column in the source
p.circle( x='x', y='y', radius='radius', color='color', legend_field='label', source=source)

labels = LabelSet(x='x', y='y', text='names',x_offset=-15,y_offset=-8,
               source=source, render_mode='canvas',text_font_size="9px",text_color="white")

hover=HoverTool()
hover.tooltips="""
<div>
<div><strong>Type  : </strong>@v</div>
<div><strong>Count : </strong>@c</div>
</div>
"""
p.add_tools(hover)
p.axis.visible = None
p.xgrid.visible = False
p.ygrid.visible = False
p.title.text = "Types of Transactions Made By User"
p.title.align = "center"
p.title.text_font_size = "15px"
output_notebook()
p.add_layout(labels)
show(p)
```
