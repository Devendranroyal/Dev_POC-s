## count of 0 nd 1 in default column using bar chart.(bar chart)

## count of Ages in age column(using bar chart).

```python
age_col_value=list(df['age'].value_counts(sort=False).index.astype(str))
```

```python

age_col_value=list(df['age'].value_counts(sort=False).index.astype(str))
counts=list(df['age'].value_counts(sort=False).values)
d=len(counts)
plot2 = figure(x_range=age_col_value[0:3], plot_height=500,title='Count of Number of Ages',
           x_axis_label='Ages', y_axis_label='Count_of_ages',
           toolbar_location='right')
plot2.vbar(x=age_col_value, top=counts, width=0.3,fill_color=Category10[3],line_color="white")
callback_range_slider = CustomJS(args=dict(fig=plot2, xr=age_col_value), code="""
    var A = slider.value[0];
    var B = slider.value[1];
    fig.x_range.factors = [];
    for (i = A; i <= B; i++) {
      fig.x_range.factors.push(xr[i-20])
    }
""")

plot2.y_range.start = 0
hover=HoverTool(tooltips=([('age','@x'),('count','@top')]))
plot2.add_tools(hover)
range_slider = RangeSlider(start=20, end=56, value=(20,22), step=1, title="Age Range Slider")
callback_range_slider.args["slider"]=range_slider
plot2.x_range.js_on_change('factors',callback_range_slider)
plot2.xgrid.visible = False
plot2.ygrid.visible = False
layout2 = column(plot2, widgetbox(range_slider))
output_notebook()
show(plot2)
```

## Plot the pie chart for percentage of 0 nd 1 using default column.

```python

xy=dict(df['default'].value_counts())
data1=pd.Series(xy).reset_index(name='counts').rename(columns={'index':'default_col_value'})

data1['percentage'] = data1['counts']/10
data1
```

```python

#data=xy=pd.Series(df['default']).value_counts()
tooltipss=[("default value:", "@default_col_value"),
  ("percentage", "@percentage")]
plot6 = figure(plot_height=500, toolbar_location='right',tooltips=tooltipss, title='Percentage of 0 and 1')
data1['angle'] = data1['percentage']/data1['percentage'].sum() * 2*np.pi
data1['color'] = ['orange', 'red']
plot6.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white",fill_color='color', source=data1)
data1['label'] = ["{:.0f}".format(l) for l in data1['percentage']]
data1['label'] = data1['label'].astype(str)

source = ColumnDataSource(data1)
#
labels = LabelSet(x=0,y=1,text='label', source=source,angle=cumsum('angle', include_zero=True),
                  text_font_size="12pt", text_color="black")
plot6.add_layout(labels)
plot6.xgrid.visible = False
plot6.ygrid.visible = False
output_notebook()
show(plot6)
```
