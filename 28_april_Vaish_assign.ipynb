### Matplotlib
It is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. It can be used in Python scripts, the Python and IPython shells, the Jupyter notebook, web application servers, and four graphical user interface toolkits.

#### matplotlib Advantages:
There are several advantages of using matplotlib to visualize data.

    - A multi-platform data visualization tool built on the numpy and sidepy framework. Therefore, it's fast and efficient.
    - It possesses the ability to work well with many operating systems and graphic backends.
    - It possesses high-quality graphics and plots to print and view for a range of graphs such as histograms, bar charts, pie charts, scatter plots and heat maps.
    - With Jupyter notebook integration, the developers have been free to spend their time implementing features rather than struggling with  compatibility.
    - It has large community support and cross-platform support as it is an open source tool.
    - It has full control over graph or plot styles such as line properties, thoughts, and access properties.


### Bokeh
It is an interactive visualization library that targets modern web browsers for presentation. Its goal is to provide elegant, concise construction of versatile graphics, and to extend this capability with high-performance interactivity over very large or streaming datasets. It can help anyone who would like to quickly and easily create interactive plots, dashboards, and data applications.

#### Bokeh  advantages:
    - Bokeh allows you to build complex statistical plots quickly and through simple commands
    - Bokeh provides you output in various medium like html, notebook and server
    - We can also embed Bokeh visualization to flask and django app
    - Bokeh can transform visualization written in other libraries like matplotlib, seaborn, ggplot
    - Bokeh has flexibility for applying interaction, layouts and different styling option to visualization
    
    
## Example:

### implementing a line plot using matplotlib and bokeh

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[4,5,7,9,11]
y=[4,6,2,4,3]
plt.plot(x,y)
plt.title("Matplotlib line plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
from bokeh.plotting import figure,output_notebook,show
x=[4,5,7,9,11]
y=[4,6,2,4,3]
output_notebook()
p=figure(title="Bokeh line plot",x_axis_label="X values",y_axis_label="Y values")
p.line(x,y,legend_label="test",line_width=3)
show(p)
```

## Wordcloud

Word clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

A word cloud is a collection, or cluster, of words depicted in different sizes. The bigger and bolder the word appears, the more often itâ€™s mentioned within a given text and the more important it is.

Also known as tag clouds or text clouds, these are ideal ways to pull out the most pertinent parts of textual data, from blog posts to databases. They can also help business users compare and contrast two different pieces of text to find the wording similarities between the two.

### Where Word Clouds Excel for Businesses

In the right setting, word cloud visualizations are a powerful tool. Here are a few instances when word clouds excel:

Finding customer pain points â€” and opportunities to connect. Do you collect feedback from your customers? (You should!) Analyzing your customer feedback can allow you to see what your customers like most about your business and what they like least. Pain points (such as â€œwait time,â€ â€œprice,â€ or â€œconvenienceâ€) are very easy to identify with text clouds.
Understanding how your employees feel about your company. Text cloud visualization can turn employee feedback from a pile of information youâ€™ll read through later to an immediately valuable company feedback that positively drives company culture.
Identifying new SEO terms to target. In addition to normal keyword research techniques, using a word cloud may make you aware of potential keywords to target that your site content already uses.

## count the trending failed workflow

```python
df1=pd.read_excel("FL.xlsx")
```

```python
df1
```

```python
df2=df1
```

```python
df2.rename(columns = {'Failed Workflow Name':'Failed_Workflow_Name'}, inplace = True)
df2.rename(columns = {'Duration of each workflow':'Duration_of_each_workflow'}, inplace = True)
```

```python
from bokeh.plotting import figure,output_notebook,show,ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10
source=ColumnDataSource(df1)
work_list=source.data['Failed_Workflow_Name'].tolist()
p=figure(y_range=work_list,plot_width=800,plot_height=600,title=" Failed work flow with Duration ",x_axis_label=" Duration (Days) ",tools="pan,box_select,xzoom_out,save,reset")
p.hbar(y='Failed_Workflow_Name',right='Duration_of_each_workflow',left=0,height=0.4,fill_color=factor_cmap('Failed_Workflow_Name',palette=Category10[10],factors=work_list),fill_alpha=0.9,source=source)
hover=HoverTool()
hover.tooltips="""
<div>
<h4>@Failed_Workflow_Name</h4>
<div><strong>Duration : </strong>@Duration_of_each_workflow</div>
</div>
"""
p.add_tools(hover)
p.x_range.start = 0
output_notebook()
show(p)
```

```python
from bokeh.plotting import figure,output_notebook,show,ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20
source1=ColumnDataSource(df3)
per_list=source1.data['Name'].tolist()
p1=figure(x_range=per_list,background_fill_color='#F0F8FF',title="Percentage Of Contribution",y_axis_label="Percentage(out of 100)",tools="pan,box_select,xzoom_out,save,reset")
p1.vbar(x='Name',top='Percentage',width=0.6,fill_color=factor_cmap('Name',palette=Category20[18],factors=per_list),fill_alpha=0.9,source=source1)
hover=HoverTool()
hover.tooltips="""
<div>
<div><strong>Contribution : @Percentage %</div>
</div>
"""
p1.add_tools(hover)
p1.outline_line_color = "navy"
p1.y_range.start = 0
p1.xaxis.major_label_orientation = "vertical"
p1.plot_width=900
output_notebook()
show(p1)
```

```python
from bokeh.plotting import figure,output_notebook,show,ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20c
source2=ColumnDataSource(df4)
time_list=source2.data['Failed_Workflow_Name'].tolist()
p2=figure(x_range=time_list,background_fill_color='white',title="Recovery time in hours",y_axis_label="recovery time(in hours)",tools="pan,box_select,xzoom_out,save,reset")
p2.vbar(x='Failed_Workflow_Name',top='Working_hours',width=0.6,fill_color=factor_cmap('Failed_Workflow_Name',palette=Category20c[10],factors=time_list),fill_alpha=0.9,source=source2)
hover=HoverTool()
hover.tooltips="""
<div>
<h3>@Failed_Workflow_Name</h3>
<div><strong>Hourstaken : @Working_hours</div>
<div><strong>Average Hours : 27.8</div>
</div>
"""
p2.add_tools(hover)
p2.outline_line_color = "navy"
p2.y_range.start = 0
p2.xaxis.major_label_orientation = "vertical"
p2.plot_width=800
output_notebook()
show(p2)
```

```python
a=df4['Failed_Workflow_Name'].tolist()
b=df4['Working_hours'].tolist()
xy=dict(zip(a, b)) 
data=pd.Series(xy).reset_index(name='Working_hours').rename(columns={'index':'Failed_Workflow_Name'})
data["Failed_Workflow_Name"]= data["Failed_Workflow_Name"].astype(str)
data
```

```python
from math import pi
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.palettes import Category20c
from bokeh.plotting import figure,ColumnDataSource
from bokeh.transform import cumsum
from bokeh.models.tools import HoverTool
from bokeh.models import Legend, HoverTool, LabelSet
data['angle'] = data['Working_hours']/data['Working_hours'].sum() * 2*pi
data['color'] = Category20c[10]
source = ColumnDataSource(data=data)
p5 = figure(background_fill_color="white", title="Recovery time in hrs", toolbar_location='right',plot_width=1200,plot_height=600,
            x_range=(-0.5, 1.0),x_axis_label='0s and 1s', y_axis_label='Count')
p5.wedge(x=0, y=1, radius=0.3,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='Failed_Workflow_Name', source=source)
hover4=HoverTool()
hover4.tooltips="""
<div>
<h3>@Failed_Workflow_Name</h3>
<div>@Working_hours</div>
<div><strong>Average Hours : 27.8</div>
</div>
"""
#labels = LabelSet(x=0, y=1,text="label", x_offset=0,
#         text_font_size="15pt", text_color="black",
#          source=source,angle=cumsum('angle', include_zero=True),text_align='left')
#p5.add_layout(labels)
p5.add_tools(hover4)
p5.axis.visible=False
p5.grid.grid_line_color = None
p5.outline_line_color = "navy"
output_notebook()
show(p5)
```
